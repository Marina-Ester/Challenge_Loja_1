### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# prompt: Nesta primeira análise, você deve calcular o faturamento total de cada loja. Somando os valores da coluna Preço de cada loja para estimar o faturamento.

faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento total Loja 1: R$ {faturamento_loja1:.2f}")
print(f"Faturamento total Loja 2: R$ {faturamento_loja2:.2f}")
print(f"Faturamento total Loja 3: R$ {faturamento_loja3:.2f}")
print(f"Faturamento total Loja 4: R$ {faturamento_loja4:.2f}")
```

# 5. Frete Médio por Loja


```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate total revenue for each store
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Create a DataFrame for plotting
faturamento_total = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento Total': [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
})

plt.figure(figsize=(8, 5))
plt.bar(faturamento_total['Loja'], faturamento_total['Faturamento Total'], color='dodgerblue')
plt.title('Faturamento Total por Loja')
plt.xlabel('ID da Loja')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Conseguimos ver que a Loja 1 vendeu mais, mas ambas com pouca diferença de vendas.

```python
# prompt: # Contagem de produtos por categoria

# Combine 'Categoria do Produto' from all dataframes
all_categories = pd.concat([loja['Categoria do Produto'], loja2['Categoria do Produto'], loja3['Categoria do Produto'], loja4['Categoria do Produto']])

# Contagem de produtos por categoria
distribuicao_categoria = all_categories.value_counts()

plt.figure(figsize=(7,7))
plt.pie(distribuicao_categoria, labels=distribuicao_categoria.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Produtos por Categoria')
plt.axis('equal')
plt.show()
```

Aqui conseguimos perceber quais são os produtos mais vendiso por categorias, Moveis e eltronicos diparam nas vendas.

```python
# Gráfico de dispersão entre nota e valor do frete

# Combine 'Frete' and 'Avaliação da compra' from all dataframes
pedidos_combined = pd.concat([loja[['Frete', 'Avaliação da compra']], loja2[['Frete', 'Avaliação da compra']], loja3[['Frete', 'Avaliação da compra']], loja4[['Frete', 'Avaliação da compra']]])

plt.figure(figsize=(8, 5))
plt.scatter(pedidos_combined['Frete'], pedidos_combined['Avaliação da compra'], alpha=0.5, color='purple')
plt.title('Avaliação dos Clientes x Valor do Frete')
plt.xlabel('Valor do Frete (R$)')
plt.ylabel('Nota do Cliente')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

Mostra se o frete mais caro pode estar influenciando a nota dos clientes.
